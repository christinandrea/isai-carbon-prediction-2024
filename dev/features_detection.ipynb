# Code Trial

The purpose of this section is to find the best parameters (clip limit and tile grid size) in CLAHE for enhance the image's quality. It will later be implemented in FAST and SIFT features detection.

```python
# Check image quality

img_path = os.path.join(hst_1_100_20_emp[3],img_hst1_100_20[3])

print(img_path)


image = cv2.imread(img_path)

# Get dimensions
height, width, _ = image.shape
print("Image resolution: {}x{}".format(width, height))


gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Compute Laplacian gradient
laplacian = cv2.Laplacian(gray, cv2.CV_64F).var()
print("Image sharpness (Laplacian):", laplacian)

# Compute standard deviation
std_dev = gray.std()
print("Image contrast (Std Deviation):", std_dev)

variance = cv2.meanStdDev(gray)[1]
print("Image noise (Variance):", variance)
```

```python
from matplotlib.cm import ScalarMappable
from skimage import metrics

trial_img = os.path.join(hst_0_25_30_emp[0],img_hst0_25_30[3])
trial_read = cv2.imread(trial_img, cv2.IMREAD_GRAYSCALE)

laplacian_ori = cv2.Laplacian(trial_read, cv2.CV_64F).var()
print("(Original Image sharpness (Laplacian):", laplacian_ori)

std_dev_ori = trial_read.std()
print("Original Image contrast (Std Deviation):", std_dev_ori)

variance_ori = cv2.meanStdDev(trial_read)[1]

print("Original Image noise (Variance):", variance_ori)


img_no_change = cv2.imread(trial_img, cv2.IMREAD_UNCHANGED)

image_depth = img_no_change.dtype

if image_depth == 'uint8':
    print("Image is 8-bit.")
elif image_depth == 'uint16':
    print("Image is 16-bit.")
else:
    print("Image depth is neither 8-bit nor 16-bit.")

grayscale_img_depth = trial_read.dtype

if grayscale_img_depth == 'uint8':
    print("Grayscale image is 8-bit.")
elif grayscale_img_depth == 'uint16':
    print("Grayscale image is 16-bit.")
else:
    print("Grayscale image depth is neither 8-bit nor 16-bit.")


print ("Image Path :", trial_img)

cliplimit = float(input("Clip limit : "))
gridSize = int(input("Num Grid Size : "))

mergeSize = (gridSize, gridSize)
clahe= cv2.createCLAHE(clipLimit=cliplimit, tileGridSize=mergeSize)
cl_img= clahe.apply(trial_read)

laplacian = cv2.Laplacian(cl_img, cv2.CV_64F).var()
print("Image sharpness (Laplacian):", laplacian)


std_dev = cl_img.std()
print("Image contrast (Std Deviation):", std_dev)

variance = cv2.meanStdDev(cl_img)[1]
print("Image noise (Variance):", variance)


kp = fast.detect(cl_img)
pts_new = [kp[x].pt for x in range(len(kp))]
np_newlist = np.array(pts_new, dtype=np.float32)

print(len(np_newlist))

print("Keypoints Detected", len(kp))
print(cl_img.shape)
print(cl_img.flatten())
print(len(cl_img.flatten()))
hist_ori, bins_ori = np.histogram(trial_read.flatten(), bins=256, range=[0, 256], density=True)
hist_clahe, bins_clahe = np.histogram(cl_img.flatten(), bins=256, range=[0, 256], density=True)

if cliplimit > 5.0:
  print("Clip limit high")
else:
  print("Clip limit low")

exceeds_limit = any(count > cliplimit for count in hist_clahe)

if exceeds_limit:
    print("At least one bin count in the histogram exceeds the clipLimit")
else:
    print("No bin count in the histogram exceeds the clipLimit")


plt.imshow(trial_read,cmap="gray")
plt.axis('off')
plt.show()

plt.imshow(cl_img,cmap="gray")
plt.axis('off')
plt.show()


ssim_index = metrics.structural_similarity(trial_read, cl_img)

print("SSIM : ", ssim_index)

print("Ori Image")
plt.figure(figsize=(8, 5))
plt.fill_between(bins_ori[:-1], hist_ori, color='skyblue', alpha=0.75)
plt.title('Histogram Density of the Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Density')
plt.show()

print("CLAHE Image")
plt.figure(figsize=(8, 5))
plt.fill_between(bins_clahe[:-1], hist_clahe, color='skyblue', alpha=0.75)
plt.title('Histogram Density of the Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Density')
plt.show()

print("Features Detected Using FAST")

img_kp = cv2.drawKeypoints(cl_img,kp,None,color=(0, 255, 0),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# img_kp = cv2.drawKeypoints(median_filtered_img, kp, None, color=(0, 255, 0), flags=0)
plt.imshow(img_kp)
plt.axis('off')
plt.show()

sift_ft = cv2.SIFT_create()
sift_kp = sift_ft.detect(cl_img,None)

sift_img = cv2.drawKeypoints(cl_img,sift_kp,None,color=(0, 255, 0),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

print("Keypoints detected usings SIFT : ",len(sift_kp))
print("Features Detected Using SIFT")
pts_new_sift = [sift_kp[x].pt for x in range(len(sift_kp))]
np_newlist_sift = np.array(pts_new_sift, dtype=np.float32)

print(sift_kp)
plt.imshow(sift_img)
plt.axis('off')
plt.show()

print(np_newlist_sift)

```
